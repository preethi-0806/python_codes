```python
import pandas as pd

# Load the dataset
data = pd.read_csv('Day_14_Pharma_data.csv')

# Check for missing values
print(data.isnull().sum())

# Check for duplicates
data = data.drop_duplicates()

# Optional: Handle missing values by dropping or filling
data = data.dropna()  # or data.fillna(value) if appropriate

import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the average effectiveness by region
effectiveness_by_region = data.groupby('Region')['Effectiveness'].mean().reset_index()

# Create a bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='Effectiveness', data=effectiveness_by_region, palette='coolwarm')
plt.title('Average Effectiveness Across Different Regions')
plt.xlabel('Region')
plt.ylabel('Average Effectiveness')
plt.xticks(rotation=45)
plt.show()

# Violin plot for Effectiveness and Side Effects
plt.figure(figsize=(12, 6))
sns.violinplot(x='Region', y='Effectiveness', data=data, inner='quart', palette='muted')
sns.violinplot(x='Region', y='Side_Effects', data=data, inner='quart', palette='muted')
plt.title('Distribution of Effectiveness and Side Effects Across Regions')
plt.xlabel('Region')
plt.ylabel('Values (Effectiveness/Side Effects)')
plt.show()

# Pairplot of Effectiveness, Side Effects, and Marketing Spend
sns.pairplot(data[['Effectiveness', 'Side_Effects', 'Marketing_Spend']], palette='Set2')
plt.suptitle('Pairplot of Effectiveness, Side Effects, and Marketing Spend', y=1.02)
plt.show()

# Boxplot of Effectiveness Across Trial Periods
plt.figure(figsize=(10, 6))
sns.boxplot(x='Trial_Period', y='Effectiveness', data=data, palette='viridis')
plt.title('Effectiveness Across Different Trial Periods')
plt.xlabel('Trial Period')
plt.ylabel('Effectiveness')
plt.show()

# Regression plot for Marketing Spend vs Effectiveness
plt.figure(figsize=(8, 6))
sns.regplot(x='Marketing_Spend', y='Effectiveness', data=data, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
plt.title('Marketing Spend vs Effectiveness')
plt.xlabel('Marketing Spend')
plt.ylabel('Effectiveness')
plt.show()
```
